### 2.1 AlexNet第一层 卷积层Conv1

1. 网络是使用两块gpu并行训练的，只用看下半部分即可。
2. 第一层卷积层Conv1，kernel_size是11，卷积核个数是48，因为分成两部分训练，所以图片中kernels还再乘2。
3. 只看下半部分，可知得到特征图大小为55×55×48(48为channel个数，和卷积核的个数相对应；55是width和height，计算公式在图片右下角，1+2的含义是图片左、上方多加了1行，右、下多加了两行，都是用0填补)

### 2.2 AlexNet第二层 池化层Maxpool1

1. 网络结构图中没有给出池化核的大小(kernel_size),也没有给步距（stride）的大小，下面是up主给的数据
2. kernel_size为3，padding为0，stride为2，根据图片右下角公式，推算出得到的特征图大小为27×27×96（注意：池化操作不会改变特征图的深度，结构图中没有给出池化之后具体的特征图）

### 2.3 AlexNet第三层 卷积层Conv2

1. 通过网络结构图只能得到，kernel_size为5，卷积核个数为128（stride和padding无法从结构图中得到，是up主给的数据）
2. 输入27×27×48，输出27×27×128

### 2.4 AlexNet第四层 池化层Maxpool2

1. 输入27×27×128，输出13×13×128，深度128不变
2. 结构图中未给padding和stride，可能论文中解释了，这里up给了，通过右下角公式可从27计算得到13
