### 2.4 分析VGG16

1. 卷积层conv的stride和padding全为1；池化层maxpool的size和stride全为2
2. 经过卷积，得到的特征图的width、height和输入特征图一样大(i.e. 224=(224-3+2)/1+1，和原来的一样)
3. 经过池化，得到的特征图的width、heighth是输入特征图的一半（i.e. 112=(224-2+0)/2+1，变为原来的一半）
